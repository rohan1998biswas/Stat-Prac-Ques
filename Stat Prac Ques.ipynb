````python
'''1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
After generating the list, find the following:
 (i) Write a Python function to calculate the mean of a given list of numbers.
 Create a function to find the median of a list of numbers.
 (ii) Develop a program to compute the mode of a list of integers.
 (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
 (iv) Write a Python function to find the geometric mean of a list of positive numbers.
 (v) Create a program to calculate the harmonic mean of a list of values.
 (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
 (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
outliers.'''
````

````python
# Ans:
import random
import statistics

int_list = [random.randint(90, 130) for _ in range(100)]

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    n = len(numbers)
    numbers.sort()
    if n % 2== 0:
        return (numbers[n//2] + numbers[n//2 - 1]) / 2
    else:
        return numbers[n//2]

def calculate_mode(numbers):
    return statistics.mode(numbers)

def calculate_weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

def calculate_geometric_mean(numbers):
    product = 1
    for num in numbers:
        product *= num
    return product ** (1 / len(numbers))

def calculate_harmonic_mean(numbers):

    harmonic_sum = 0
    for num in numbers:
        harmonic_sum += 1 / num
    return len(numbers) / harmonic_sum

def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

def calculate_trimmed_mean(numbers, trim_percent):
    n = len(numbers)
    trim_index = int(n * trim_percent / 100)
    sorted_numbers = sorted(numbers)
    trimmed_numbers = sorted_numbers[trim_index:-trim_index]
    return calculate_mean(trimmed_numbers)

mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)

weights = [random.randint(1, 5) for _ in range(100)]
weighted_mean = calculate_weighted_mean(int_list, weights)

geometric_mean = calculate_geometric_mean(int_list)
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean = calculate_trimmed_mean(int_list, 10)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Weighted Mean:", weighted_mean)
print("Geometric Mean:", geometric_mean)
print("Harmonic Mean:", harmonic_mean)
print("Midrange:", midrange)
print("Trimmed Mean (10%):", trimmed_mean)
````

````python
# Ans:
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

int_list2 = [random.randint(200, 300) for _ in range(500)]

data = np.array(int_list2)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=20, edgecolor='black')
plt.title('Frequency Distribution')
plt.subplot(1, 2, 2)
sns.distplot(data, kde=False)
x = np.linspace(min(data), max(data), 100)
y = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, y, color='red')
plt.title('Gaussian Distribution')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, stat='density')
plt.title('Frequency and Smoothed KDE Plot')
plt.show()

plt.figure(figsize=(10, 6))
sns.distplot(data, kde=True)
x = np.linspace(min(data), max(data), 100)
y = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, y, color='red')
plt.title('Gaussian Distribution and Smoothed KDE Plot')
plt.show()

def calculate_range(numbers):
    return max(numbers) - min(numbers)

def calculate_variance_std(numbers):

    variance = np.var(numbers)
    std_dev = np.std(numbers)
    return variance, std_dev

def calculate_iqr(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return q3 - q1

def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers)
    return (std_dev / mean) * 100

def calculate_mad(numbers):
    mean = np.mean(numbers)
    mad = np.mean(np.abs(numbers - mean))
    return mad

def calculate_quartile_deviation(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return (q3 - q1) / 2

def calculate_range_based_coefficient_of_dispersion(numbers):
    range_val = calculate_range(numbers)
    mean = np.mean(numbers)
    return range_val / mean

range_val = calculate_range(int_list2)
variance, std_dev = calculate_variance_std(int_list2)
iqr = calculate_iqr(int_list2)
coefficient_of_variation = calculate_coefficient_of_variation(int_list2)
mad = calculate_mad(int_list2)
quartile_deviation = calculate_quartile_deviation(int_list2)
range_based_coefficient_of_dispersion = calculate_range_based_coefficient_of_dispersion(int_list2)

print("Range:", range_val)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Interquartile Range (IQR):", iqr)
print("Coefficient of Variation:", coefficient_of_variation)
print("Mean Absolute Deviation (MAD):", mad)
print("Quartile Deviation:", quartile_deviation)
print("Range-Based Coefficient of Dispersion:", range_based_coefficient_of_dispersion)
````

````python
'''3. Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.'''
````

````python
# Ans:
class DiscreteRandomVariable:
    def __init__(self, pmf):
        self.pmf = pmf

    def expected_value(self):
        expected_value = 0
        for value, probability in self.pmf.items():
            expected_value += value * probability
        return expected_value

    def variance(self):
        expected_value = self.expected_value()
        variance = 0
        for value, probability in self.pmf.items():
            variance += probability * (value - expected_value)**2
        return variance

pmf = {1: 0.2, 2: 0.3, 3: 0.5}
rv = DiscreteRandomVariable(pmf)

print("Expected value:", rv.expected_value())
print("Variance:", rv.variance())
````

````python
'''4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes.'''
````

````python
# Ans:
def simulate_die_roll(num_rolls):
    rolls = []
    for _ in range(num_rolls):
        roll = random.randint(1, 6)
        rolls.append(roll)
    return rolls

def calculate_expected_value(rolls):
    return sum(rolls) / len(rolls)

def calculate_variance(rolls):
    mean = calculate_expected_value(rolls)
    variance = sum((roll - mean)**2 for roll in rolls) / len(rolls)
    return variance

num_rolls = 10000
rolls = simulate_die_roll(num_rolls)
expected_value = calculate_expected_value(rolls)
variance = calculate_variance(rolls)
print("Expected value:", expected_value)
print("Variance:", variance)
````

````python
# Ans:
def generate_random_samples(distribution, params, sample_size):
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size=sample_size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size=sample_size)
    else:
        raise ValueError("Invalid distribution")
    return samples

def calculate_mean_and_variance(samples):
    mean = np.mean(samples)
    variance = np.var(samples)
    return mean, variance

binomial_samples = generate_random_samples('binomial', (10, 0.5), 1000)
binomial_mean, binomial_variance = calculate_mean_and_variance(binomial_samples)
poisson_samples = generate_random_samples('poisson', (5,), 1000)
poisson_mean, poisson_variance = calculate_mean_and_variance(poisson_samples)
print("Binomial distribution:")
print("Mean:", binomial_mean)
print("Variance:", binomial_variance)
print("\nPoisson distribution:")
print("Mean:", poisson_mean)
print("Variance:", poisson_variance)
plt.hist(binomial_samples, bins=20, density=True, alpha=0.7, label='Binomial')
plt.hist(poisson_samples, bins=20, density=True, alpha=0.7, label='Poisson')
plt.legend()
plt.show()
````

````python
# Ans:
def normal_pdf(x, mean, std_dev):
  pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
  return pdf
mean = 0
std_dev = 1
x_values = np.linspace(-3, 3, 100)
pdf_values = [normal_pdf(x, mean, std_dev) for x in x_values]
plt.plot(x_values, pdf_values)
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.title('Normal Distribution PDF')
plt.show()
````

````python
# Ans:
def poisson_pmf(k, lamda):
  pmf = (lamda**k * np.exp(-lamda)) / np.math.factorial(k)
  return pmf

lamda = 3
k_values = np.arange(10)
pmf_values = [poisson_pmf(k, lamda) for k in k_values]
plt.bar(k_values, pmf_values)
plt.xlabel('k')
plt.ylabel('PMF(k)')
plt.title('Poisson Distribution PMF')
plt.show()
````

````python
# Ans:
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)
p1 = np.mean(old_layout)
p2 = np.mean(new_layout)
p_pooled = (np.sum(old_layout) + np.sum(new_layout)) / (len(old_layout) + len(new_layout))
se = np.sqrt(p_pooled * (1 - p_pooled) * (1/len(old_layout) + 1/len(new_layout)))
z_score = (p2 - p1) / se
p_value = 2 * (1 - np.abs(z_score))
print("z-score:", z_score)
print("p-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("The new layout is significantly more successful than the old layout.")
else:
    print("There is no significant difference between the two layouts.")
````

````python
# Ans:
def compare_product_versions(version1_scores, version2_scores):
    mean1 = np.mean(version1_scores)
    std1 = np.std(version1_scores, ddof=1)
    n1 = len(version1_scores)
    mean2 = np.mean(version2_scores)
    std2 = np.std(version2_scores, ddof=1)
    n2 = len(version2_scores)
    pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))
    t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))
    degrees_of_freedom = n1 + n2 - 2
    p_value = stats.t.sf(abs(t_statistic), df=degrees_of_freedom) * 2
    return t_statistic, degrees_of_freedom, p_value

version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]
t_stat, df, p_value = compare_product_versions(version1_scores, version2_scores)
print("t-statistic:", t_stat)
print("Degrees of freedom:", df)
print("p-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in quality between the two versions.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of a difference in quality.")
````

````python
'''21. A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.
 Use the below code to generate data:
 ```python
 np.random.seed(0)
 age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)
 voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
'''
````

````python
# Ans:
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)
contingency_table = np.array([
    [np.sum((age_groups == '18-30') & (voter_preferences == 'Candidate A')), np.sum((age_groups == '18-30') & (voter_preferences == 'Candidate B'))],
    [np.sum((age_groups == '31-50') & (voter_preferences == 'Candidate A')), np.sum((age_groups == '31-50') & (voter_preferences == 'Candidate B'))],
    [np.sum((age_groups == '51+') & (voter_preferences == 'Candidate A')), np.sum((age_groups == '51+') & (voter_preferences == 'Candidate B'))]
])
chi2, p_value, _, _ = stats.chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of an association between age groups and voter preferences.")
````
